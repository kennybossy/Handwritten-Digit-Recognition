#Import important library


```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
import matplotlib.pyplot as plt
```

#load the dataset and train test split it

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

#checking the shape of Dataset as its a 3D array

```python
x_test.shape
```

#printing the y train to check output as its 10 numbers

```python
y_train
```

#Converting the array into image using matplotlib file


```python
import matplotlib.pyplot as plt
plt.imshow(x_train[2])
```

#scalling the dataset into the range of 0 to 1

```python
x_train = x_train/255
x_test = x_test/25
x_train[0]
```

#Intoducing the layers of network in the structure

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

#printing the summary of model

```python
model.summary()
```

#compiling the model with some functions like optimizer , loss etc.

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

#fit the model call it with refernce to an object

```python
history = model.fit(x_train,y_train,epochs=25,validation_split=0.2)
```

#Predicting the value then taking maximum value in the specific row

```python
y_prob = model.predict(x_test)
y_pred = y_prob.argmax(axis=1)
```

#Loading the accuracy_score from sklearn to calculate the accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

#Ploting the grpah between the loss and validation loss variables for each trained dataset

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

#Again repeating the same thing for the accuray and validation accuracy

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
